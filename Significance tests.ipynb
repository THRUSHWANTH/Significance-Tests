# Terminologies 

**Test Statistic**       - A test statistic is a number calculated by a statistical test. It describes how far your observed data is from the null hypothesis of no relationship between variables or no difference among sample groups

**P - Value**            - Probability of observing the apparant effect assuming the null hypothesis is true. It is observing the change in data (as extreme) under assumption that null hypothesis is true. It is evidence against null hypothesis. 

**Confidence Interval**  - It is the probability that a population parameter will fall between a set of values.

**Population Parameter** - It is the quatitaive characteritic of a Population like mean, varience etc which we are interested in finding out.

**Significane Level(α)**    - It represents the probability of rejecting a True null hypothesis.The threshold is set before conducting the hypothesis testing.

# Significance Tests

The purpose of significance tests is to use sample statistics to draw conclusions about population parameters. Rather than memorizing each test (which can be checked online), our goal is to identify the correct test when needed.

Note: Assumptions about the data needed for the test to be valid are crucial.

**Z-Test:**
The Z-test is a statistical method used to determine if there is a significant difference between sample and population means or between the means of two samples. It is particularly applicable when the population standard deviation is known, as the Z-test is based on the standard normal distribution (Z-distribution). However, it is rarely used, as the standard deviation of the population is often difficult to determine.

**Student t-Test:**
In this test, the statistic follows a Student's t-distribution. Its most common application is to test whether the means of two populations are significantly different. The data follows a Student t-distribution with n−1 degrees of freedom.

**ANOVA Test:**
ANOVA, or Analysis of Variance, is a statistical test used to compare the means of two or more groups. The distribution of sample means is assumed to be normal, and each population has the same variance. Samples are independent and normal. The ANOVA test generates an F-statistic, and the decision to reject the null hypothesis is based on comparing the calculated F-statistic to a critical value or by calculating a p-value.

**Chi-squared Test:**
The chi-squared test is used to determine whether there is a significant association between two categorical variables. The null hypothesis is that the variables are independent, meaning there is no association between them. The test involves comparing the observed frequencies in a contingency table to the frequencies that would be expected if the variables were independent.It is valid only when the test statistic is chi-squared distributed.These are two kinds 
1. Chi-squared tests for Independence.
2. Chi-squared tests for goodness of fit.
